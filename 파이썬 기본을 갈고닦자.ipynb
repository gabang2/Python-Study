## 3. While 반복문

```python
c = 5
while c!=0:
    print(c)
    c-=1
```

```python
while True:
    print("CTRL+C로 빠져나오세요")
```

```python
while True:
    response = input("숫자를 입력하세요")
    if int(response) % 10 == 0:
        print("10으로 나누었을 때 나머지가 0 입니다.")
        break
    print("10으로 나누었을때 나머지가 0이 아닙니다.")
```

```python
while True:
    response = input('숫자를 입력하세요')
    result = int(response) % 10
    if result == 0:
        continue
    print("10으로 나눈 나머지는{}입니다.".format(result))
```

```python
result = 1
while result != 0:
	result = int(input("숫자를 입력하세요"))
	if result == 100:
		print("while문을 빠져나옵니다")
		break
	elif result == 50:
		print("다시 입력받습니다! 50은 싫어요!")
		continue
	else:
		print(result)
```

#### format 사용하기

```python
name = "유재석"
age = "42"
a = 'Name: {name}, Age: {age}인 사람입니다'.format(name=name, age=age)
print(a)
```

```python
pos = [233,444, 654]
a = 'A의 좌표는 x={p[0]}, y={p[1]}, z={p[2]} 입니다.'.format(p=pos)
print(a)
```

```python
import math
a = '수학에서 파이 = {m.pi}'.format(m = math)
print(a)
```

```python
import math
a = '수학에서 파이 = {m.pi:.3f}'.format(m = math)
print(a)
```

```python
a='abcdef'
b=a.capitalize()
print(a, b, sep="  //  ")
```

```python
a = '    gayeong       '
b = a.strip()
print(a, b, sep = "//")
```
